```python
import pandas as pd
import glob
```

```python
csv_files = glob.glob("*.csv")
print(csv_files)
```

```python
df_list = []

for file in csv_files:
    if file != "Combined_dataset.csv":
        df = pd.read_csv(file)
        df_list.append(df)

combined_df = pd.concat(df_list, ignore_index=True)

print(combined_df.head())
```

```python
print("Shape:", combined_df.shape)
```

```python
print(combined_df.columns)
```

```python
print(combined_df.dtypes)
```

```python
print(combined_df.isnull().sum())
```

```python
combined_df = combined_df.drop_duplicates()
```

```python
print(combined_df.columns)
```

```python
combined_df["total_amount"] = combined_df["final_price"] * combined_df["ratings_count"]
```

```python
combined_df.to_csv("cleaned_combined.csv", index=False)
```
